```python
import pandas as pd
from autots import AutoTS
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
print("Enter any of the crypto name('aave', 'binance', 'bitcoin', 'dogecoin', 'ethereum', 'litecoin', 'tron') and exit for exit: ")
crypto = str(input("Enter the crypto currency: "))

while crypto != 'exit':
    
    # Load the cryptocurrency data
    data = pd.read_csv(f"{crypto.capitalize()}.csv")
    
    # Fit the AutoTS model
    model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple', drop_data_older_than_periods=200)
    model = model.fit(data, date_col='Date', value_col='Close', id_col=None)

    # Make predictions
    prediction = model.predict()
    forecast = prediction.forecast
    
    # Load the actual prices for Bitcoin
    actual_prices = data["Close"].values[-len(forecast):]

    # Calculate accuracy metrics
    mae = mean_absolute_error(actual_prices, forecast)
    rmse = np.sqrt(mean_squared_error(actual_prices, forecast))

    print(f"{crypto.capitalize()} Price Prediction:")
    print(f"The predictions are: {forecast}" )
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Root Mean Squared Error (RMSE): {rmse}\n")
    
    print("Enter any of the crypto name('aave', 'binance', 'bitcoin', 'dogecoin', 'ethereum', 'litecoin', 'tron') and exit for exit: ")
    crypto = str(input("Enter the crypto currency: "))
```
